#### 1. Find the student name who scored maximum scores in all (exam, quiz and homework)?

```python
stg1={"$unwind":"$scores"}
stg2={"$group":{"_id":"$_id", "name":{"$addToSet":'$name'}, "total_Marks":{"$sum":"$scores.score"}}}
stg3={"$sort":{"total_Marks":-1}}
stg4={"$limit":1}
for i in student_collection.aggregate([stg1,stg2,stg3,stg4]):
    print(i)
```

#### 2.Find students who scored below average in the exam and pass mark is 40%?


```python
stg1={"$unwind":"$scores"}
stg2 = {"$match":{"$and":[{"scores.type":{"$in":["exam"]} },{"scores.score":{"$gt":40}} ,{"scores.score":{"$lte":52.3}}   ]}}

for x in student_collection.aggregate([stg1,stg2]):
    print(x)
```

#### 3. Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.


```python
def agri(student):
 pCent = student.aggregate([{'$unwind':'$scores'},
                          {'$group':{"_id":"$_id","name":{"$first":"$name"},"max_score":{'$sum':"$scores.score"}}},
                          {'$addFields':{"percentage":{'$divide':["$max_score",3]}}},
                           {'$addFields':{"status":{"$cond":{'if' :{"$gte":["$percentage",40]},'then':"Pass",'else':"Fail"}}}}])
 return pCent

pCent=agri(student_collection)
for i in pCent:
 if i['percentage']>=40 and i['percentage']<=55:
     print(i)

pCent=agri(student_collection)
for i in pCent:
  print(i)
```

#### 4. Find the total and average of the exam, quiz and homework and store them in a separate collection.


```python
my_db.create_collection('total_average')
x2 = student_collection.aggregate([{'$unwind':'$scores'},{'$group':{'_id':{'type':"$scores.type"},"total:":{'$sum':"$scores.score"},"average:":{'$avg':"$scores.score"}}}])
t_a=my_db.total_average
for i in x2:
  print(i)
  t_a.insert_one(i)
```
